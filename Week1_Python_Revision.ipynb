## 1. Variables and basic data types

A variable is just a name we give to a piece of information so we can use it again later. Think of it as a labelled box. You put something inside the box, and from then on you refer to the box by its label instead of repeating the value every time.

Python has a few basic types of information you will use constantly.

**Integer**, a whole number, such as 5 or minus 3.

**Float**, a number with a decimal point, such as 5.0 or 3.14.

**String**, a piece of text, written inside quotes, such as "hello".

**Boolean**, a value that is either True or False.

You do not need to tell Python what type a variable is. It figures that out from the value you give it.

```python
# Examples of variables and their types

number_of_qubits = 3
circuit_depth = 5.0
backend_name = "Qiskit Simulator"
is_simulation = True

print(number_of_qubits)
print(circuit_depth)
print(backend_name)
print(is_simulation)
```

You can check the type of any variable using the `type` function. This is useful when you are debugging and not sure what kind of value you are working with.

```python
print(type(number_of_qubits))
print(type(circuit_depth))
print(type(backend_name))
print(type(is_simulation))
```

You can do arithmetic on numbers the same way you would on a calculator.

```python
shots = 1024
repetitions = 3

total_shots = shots * repetitions
print(total_shots)

average_error = 0.02
adjusted_error = average_error / 2
print(adjusted_error)
```

## 2. Lists

A list is a way to store many values under one name, in a specific order. Think of it as a single box that contains several smaller items lined up inside it.

Lists are written with square brackets, and the items inside are separated by commas.

```python
qubit_labels = ["q0", "q1", "q2"]
measurement_results = [0, 1, 1, 0, 1]

print(qubit_labels)
print(measurement_results)
```

Each item in a list has a position, called an index. Python starts counting positions from 0, not from 1. So the first item is at position 0, the second item is at position 1, and so on.

```python
print(qubit_labels[0])
print(qubit_labels[1])
print(qubit_labels[2])
```

You can also find out how many items are in a list using the `len` function, and you can add a new item to the end of a list using `append`.

```python
number_of_qubits_in_list = len(qubit_labels)
print(number_of_qubits_in_list)

qubit_labels.append("q3")
print(qubit_labels)
```

## 3. Conditionals

A conditional lets your code make a decision. It checks whether something is True or False, and runs different code depending on the answer.

The keyword `if` starts the check. The keyword `else` tells Python what to do if the check fails. Notice that the line after `if` and `else` is indented. In Python, indentation is not just for readability, it tells Python which lines belong inside the conditional.

```python
error_rate = 0.05
threshold = 0.1

if error_rate < threshold:
    print("Error rate is acceptable")
else:
    print("Error rate is too high")
```

You can also chain multiple conditions using `elif`, which means "else if". Python checks each condition from top to bottom and stops at the first one that is True.

```python
fidelity = 0.91

if fidelity > 0.95:
    print("Excellent result")
elif fidelity > 0.85:
    print("Good result")
else:
    print("Needs improvement")
```

## 4. Loops

A loop lets your code repeat an action without you writing the same line over and over. The most common loop in Python is the `for` loop, which goes through each item in a list, one at a time.

```python
qubit_labels = ["q0", "q1", "q2"]

for label in qubit_labels:
    print(label)
```

Here, `label` is a temporary name we chose for whatever item the loop is currently looking at. You can call it anything you like, but a clear name makes the code easier to read.

Sometimes you do not want to loop over a list, you just want to repeat something a fixed number of times. For that, we use `range`.

```python
for i in range(5):
    print("Repetition number", i)
```

Notice that `range(5)` gives you the numbers 0, 1, 2, 3, 4. It stops before reaching 5. This trips up almost everyone the first time, so it is worth remembering.

You can combine a loop with a conditional. This is one of the most common patterns you will use.

```python
measurement_results = [0, 1, 1, 0, 1, 1, 1]

count_of_ones = 0

for result in measurement_results:
    if result == 1:
        count_of_ones = count_of_ones + 1

print("Number of ones measured:", count_of_ones)
```

## 5. Functions

A function is a small, named block of code that you can reuse. Instead of writing the same calculation many times, you write it once inside a function, give it a name, and then call that name whenever you need it.

A function is defined using the keyword `def`, followed by the function name and parentheses. Anything inside the parentheses is an input the function expects. The `return` keyword sends a result back out of the function.

```python
def calculate_total_shots(shots_per_circuit, number_of_circuits):
    total = shots_per_circuit * number_of_circuits
    return total

result = calculate_total_shots(1024, 4)
print(result)
```

Once a function is defined, you can call it as many times as you like with different inputs, and it will give you a fresh result each time.

```python
print(calculate_total_shots(512, 2))
print(calculate_total_shots(2048, 10))
```

Functions are useful because they let you describe an idea once and reuse it everywhere, instead of repeating logic across your notebook. This becomes important once our circuits and calculations get longer.
