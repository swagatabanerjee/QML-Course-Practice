## Question 1, Complex Numbers

Let

$$z = 5 - 2i$$

Using NumPy, compute and print the real part, the imaginary part, the modulus, and the conjugate of $z$.

Then verify the identity

$$|z|^2 = z z^*$$

by computing both sides and printing them.

```python
# Your code here
z = 5 - 2j
real_part = np.real(z)
imag_part = np.imag(z)
z_conjugate= np.conj(z)
print("real_part: ", real_part)
print("imag_part: ", imag_part)
print("z_conjugate: ", z_conjugate)

modulus = np.abs(z) ** 2
product=z*z_conjugate
proof = np.allclose(modulus, product)
print("modulus: ",modulus)
print("product: ", product)
print("proof: ",proof)
```

## Question 2, Vectors and Kets

Define the two basic qubit states

$$|0\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix}, \qquad |1\rangle = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$$

as NumPy arrays.

Using these, construct the state

$$|-\rangle = \frac{1}{\sqrt{2}} \big(|0\rangle - |1\rangle\big)$$

Print the resulting vector, then compute and print its norm. The norm should be very close to one.

```python
# Your code here
ket_zero= np.array([1,0])
ket_one= np.array([0,1])
normalization_factor= 1/np.sqrt(2)

ket_minus = normalization_factor * (ket_zero - ket_one)

print("ket_minus: ",ket_minus)

norm_of_ket_minus = np.linalg.norm(ket_minus)
print("norm_of_ket_minus", norm_of_ket_minus)

proof= np.allclose(1,norm_of_ket_minus)

print("norm close to 1? ", proof)
```

## Question 3, Inner Products

Let

$$u = \begin{pmatrix} 2 - i \\ 3 \end{pmatrix}, \qquad v = \begin{pmatrix} 1 \\ 2 + i \end{pmatrix}$$

Compute $\langle u, v \rangle$ using the correct NumPy function for a quantum mechanical inner product.

Then check whether $u$ and $v$ are orthogonal.

```python
# Your code here
u= np.array([ 2-1j , 3 ])
v= np.array([ 1 , 2+1j ])

inner_product=  np.vdot(u,v)
print(inner_product)
is_orthogonal= np.allclose(0,inner_product)
print("is u and v orhogonal: ", is_orthogonal)
```
